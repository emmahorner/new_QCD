```python
dataQCD = np.load('3to1 model/final3to1x1.20e-03x1.60e-09.npz') 
dataDM = np.load('../dm_sterile_production/3to1 model/final3to1x1.20e-03x1.60e-09.npz')
```

```python

## total
plt.figure()
plt.plot(dataQCD['epsilon'],dataQCD['epsilon']**2 * dataQCD['final_distribution']*2000, color = '#5CBE00')
plt.plot(dataDM['epsilon'],dataDM['epsilon']**2 * dataDM['final_distribution']*2000, color = '#0B3760', linestyle = '--')

plt.legend(['QCD', 'DM'], fontsize=10)
plt.xlabel(r"$\epsilon = p_\nu / T_{cm}$", fontsize=10)
plt.ylabel(r"$\nu_s$ distribution (arb. units)", fontsize=10)
plt.title('$f_{s} +\overline{f_{s}}$')

## sterile evo
plt.figure()
plt.plot(dataQCD['epsilon'],dataQCD['epsilon']**2 * dataQCD['sterile_evo'][-1]*2000, color = '#5CBE00')
plt.plot(dataDM['epsilon'],dataDM['epsilon']**2 * dataDM['sterile_evo'][-1]*2000, linestyle = '--', color = '#0B3760')
plt.legend(['QCD', 'DM'], fontsize=10)
plt.xlabel(r"$\epsilon = p_\nu / T_{cm}$", fontsize=10)
plt.ylabel(r"$\nu_s$ distribution (arb. units)", fontsize=10)
plt.title('$f_{s}$')

## anti evo
plt.figure()
plt.plot(dataQCD['epsilon'],dataQCD['epsilon']**2 * dataQCD['anti_sterile_evo'][-1]*2000, color = '#5CBE00')
plt.plot(dataDM['epsilon'],dataDM['epsilon']**2 * dataDM['anti_sterile_evo'][-1]*2000, linestyle = '--', color = '#0B3760')
plt.legend(['QCD', 'DM'], fontsize=10)
plt.xlabel(r"$\epsilon = p_\nu / T_{cm}$", fontsize=10)
plt.ylabel(r"$\nu_s$ distribution (arb. units)", fontsize=10)
plt.title('$\overline{f_{s}}$')
```

```python
#electron
plt.figure()
plt.semilogx(dataQCD['T'], 10**4*dataQCD['lep_evo_e'], color = '#5CBE00')
plt.semilogx(dataDM['T'], 10**4*dataDM['lep_evo_e'], color = '#008185')
plt.legend(['QCD', 'DM'], fontsize=10)
plt.xlim(2000,10)
plt.xlabel("T (MeV)")
plt.ylabel(r"$L_e$ Number $(\times 10^{-4})}$")
plt.title('$L_e$')


#muon
plt.figure()
plt.semilogx(dataQCD['T'], 10**4*dataQCD['lep_evo_mu'], color = '#5CBE00')
plt.semilogx(dataDM['T'], 10**4*dataDM['lep_evo_mu'], color = '#008185')
plt.legend(['QCD', 'DM'], fontsize=10)
plt.xlim(2000,10)
plt.xlabel("T (MeV)")
plt.ylabel(r"$L_{\mu}$ Number $(\times 10^{-4})}$")
plt.title('$L_{\mu}$')

#tau
plt.figure()
plt.semilogx(dataQCD['T'], 10**4*dataQCD['lep_evo_tau'], color = '#5CBE00')
plt.semilogx(dataDM['T'], 10**4*dataDM['lep_evo_tau'], color = '#008185')
plt.legend(['QCD', 'DM'], fontsize=10)
plt.xlim(2000,10)
plt.xlabel("T (MeV)")
plt.ylabel(r"$L_{\tau}$ Number $(\times 10^{-4})}$")
plt.title('$L_{\tau}$')
```
